### 3.4 Seperating columns with respect to data types (Num, Cat and Bool).

- Firstly, data should be seperated according to the dtypes before begining with preprocessing of the data.
- This helps in treating the missing values accordingly.

```python
#Code to seperate the Dataset on the basis of 'dtypes'
hdN=pd.DataFrame()
hdB=pd.DataFrame(hd['CentralAir'])
hdC=pd.DataFrame()

for i in hd.columns:
    if (hd[i].dtype=='int64') | (hd[i].dtype=='float64'):
        hdN[i]=hd[i]
    elif (hd[i].dtype=='object') & (i!='CentralAir'):
        hdC[i]=hd[i]
```

### 3.5 Copying columns from Numerical Column to Categrical.
- The list of columns with years information is to be copied to categorical set(hdC).
- because, when replacing NaN values with the mean values during the treatment. 
- Missing or Nan Values in the Years involved columns would be filled with the mean.

```python
Lst=['YearBuilt','YearRemodAdd','MoSold','YrSold'] #List of columns which are categorical but, present in Numerical Data.

#Adding the Lst columns from Numerical(hdN) to Categorical(hdC) Data
for i in Lst:
    hdC[i]=hdN[i]


#Dropping Lst columns from Numerical(hdN) Data
hdN=hdN.drop(Lst,axis=1)
```

```python
#Replacing Missing Values with the Mean 

for i in hdN.columns:
    if((hdN[i].count())!=1460):
        hdN[i]= hdN[i].replace(np.NaN , hdN[i].mean())

#hdN.isnull().sum() 
```

### 4.4 Outliers Detection and Handling
- Treating the outliers of the data using Inter Quantile Range.
- IQR is one of the most effective techniques in removing the outliers.
- The values over the quantile range are treated as outliers and removed.

```python
#Function for Outlier removal using Inter Qunatile Range
def OutlrRem(df,col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1
    LwrR = df[col].min()
    UprR = Q3 + (1.5*IQR)
    df = df.loc[(df[col]>LwrR)&(df[col]<UprR)]
    return df
```

## 6. Data Preparation for Model Input

### 6.1 Scaling the Data

- The values in each column is of different scales.
- And this can lead to problems when fitting.
- Scaling will help to Normalize data to single scaler.
- Here, StandardScaler is used for scaling.

```python
from sklearn.preprocessing import StandardScaler
def scaling(x):
    sclr = StandardScaler().fit(x)  #Fitting the data 
    hdtf = sclr.transform(x)        #Transforming the data
    return hdtf

def postscaling(y,z):
    hdSc=pd.DataFrame(y)
    hdSc.columns = z.columns
    return hdSc
```

```python
hdSc = postscaling(scaling(hdM),hdM) #Executing the function to check for errors.
```

### 6.2 Splitting the Data

- Functions for splitting the Data into X and Y before doing Train-Test Split

```python
def splitX(hdSc):
    X = pd.DataFrame()
    X = hdSc.loc[:,hdSc.columns!='SalePrice']
    return X

def splitY(hdSc):
    Y = pd.DataFrame()
    Y = hdSc['SalePrice']
    return Y
    
```

### 7.1 Linear Regression Model

### 7.1.1 Linear Regression Deafult Model
- Linear Regression is choosen as a primary model for train and prediction.
- The model below is basic meaning without any parameters.

```python
from sklearn.linear_model import LinearRegression
def lnrReg(X_train, Y_train, X_test):
    lr = LinearRegression()                  #Instantiating the Model
    lr = lr.fit(X_train, Y_train)            #Training the Model with X_Train and Y_Train
    Y_pred_train = lr.predict(X_train)       #Prediction on X_Train
    Y_pred_test = lr.predict(X_test)         #Prediction on Y_Train
    return Y_pred_train,Y_pred_test
```

### 7.1.2 Linear Regression GridSearchCV Model
- Linear Regression with GridSearchCV for train and prediction.
- The model below would perform GridSearchCV within the parameters provided such as, **{'normalize':[True,False], 'copy_X':[True,False],'fit_intercept':[True,False]}**
- And then, considers the best estimator for predictions.

```python
from sklearn.model_selection import GridSearchCV

def LinGridCV(X_train, Y_train,X_test):
    parameters = {'normalize':[True,False], 'copy_X':[True,False],'fit_intercept':[True,False]}
    lr = LinearRegression()
    lr = GridSearchCV(lr,parameters,cv=15)
    lr.fit(X_train, Y_train)
    
    lin = lr.best_estimator_
    print('Best Linear Regressor GSCV Estimator is',lin)
    lin = lin.fit(X_train, Y_train)
    Y_pred_train_LinGCV = lin.predict(X_train)
    Y_pred_test_LinGCV = lin.predict(X_test)
    return Y_pred_train_LinGCV,Y_pred_test_LinGCV
```

## 7.2 Decision Tree Regressor

### 7.2.1 Decision Tree Regressor Default Model
- DecisionTree Regressor is choosen as a secondary model for train and prediction.
- The model below is basic meaning without any parameters.

```python
from sklearn.tree import DecisionTreeRegressor
def DecTreReg(X_train, Y_train,X_test):
    dtr = DecisionTreeRegressor(random_state=3)
    dtr = dtr.fit(X_train, Y_train)
    Y_pred_train_dtr = dtr.predict(X_train)
    Y_pred_test_dtr = dtr.predict(X_test)
    return Y_pred_train_dtr,Y_pred_test_dtr
```

### 7.2.2 Decision Tree Regressor Grid SearchCV Model
- DecisionTree Regressor with GridSearchCV for train and prediction.
- The model below would perform GridSearchCV within the parameters provided such as, **{'criterion':['mse','mae'],'max_depth': range(2,20),'max_features': ['auto', 'sqrt', 'log2', None]}**
- And then, considers the best estimator for predictions.

```python
def dtrGridCV(X_train, Y_train,X_test):
    parameters = {'criterion':['mse','mae'],'max_depth': range(2,20),
                             'max_features': ['auto', 'sqrt', 'log2', None]}
    dt = DecisionTreeRegressor()
    dt = GridSearchCV(dt,parameters, cv=10)
    dt.fit(X_train, Y_train)
    
    dct = dt.best_estimator_
    print('Best Decision Tree GSCV Estimator is',dct)
    dct = dct.fit(X_train, Y_train)
    Y_pred_train_dtrGCV = dct.predict(X_train)
    Y_pred_test_dtrGCV = dct.predict(X_test)
    return Y_pred_train_dtrGCV,Y_pred_test_dtrGCV
```

## 7.3 Random Forest Regressor

### 7.3.1 Random Forest Regressor Default Model
- Random Forest Regressor with default parameters for train and prediction.

```python
from sklearn.ensemble import RandomForestRegressor
def RanForReg(X_train, Y_train,X_test):
    rfr = RandomForestRegressor(random_state=4)
    rfr = rfr.fit(X_train, Y_train)
    Y_pred_train_rfr = rfr.predict(X_train)
    Y_pred_test_rfr = rfr.predict(X_test)
    return Y_pred_train_rfr,Y_pred_test_rfr
```

### 7.3.2 Random Forest Regressor GridSearchCV Model
- Random Forest Regressor with GridSearchCV for train and prediction.
- The model below would perform GridSearchCV within the parameters provided such as, **{'n_estimators':[50,100,150,200],'criterion':['mse', 'mae'],'max_depth': range(3,8),'min_samples_leaf':range(2,4),'max_features': ['auto', 'sqrt', 'log2', None]}**
- And then, considers the best estimator for predictions.

```python
def rfrGridCV(X_train, Y_train,X_test):

    param_grid = {'n_estimators':[50,100,150,200],'criterion':['mse', 'mae'],'max_depth': range(3,8),
                  'min_samples_leaf':range(2,4),'max_features': ['auto', 'sqrt', 'log2', None]}
    rf = RandomForestRegressor(random_state=5)
    rf = GridSearchCV(estimator=rf,param_grid=param_grid, cv=10)
    rf.fit(X_train, Y_train)
    
    rfc = rf.best_estimator_
    print('Best Random Forest GSCV Estimator is',rfc)
    rfc = rfc.fit(X_train, Y_train)
    Y_pred_train_rfcGCV = rfc.predict(X_train)
    Y_pred_test_rfcGCV = rfc.predict(X_test)
    return Y_pred_train_rfcGCV,Y_pred_test_rfcGCV
```

## 8. Metrics for Model Performance

- Function to iterate over the above models and predict values.
- The predicted values are stored into Data Frame for comparision.
- MAE, MSE, RMSE and R-Squared metrics are used to measure the performance of the model.
- Not only the predictions on Test but, Predictions on Train are also considered for better insights.

```python
def Modl(x):
    
    #Scaling
    hdSc = postscaling(scaling(x),x)
    
    
    X = splitX(hdSc)  #X Split
    Y = splitY(hdSc)  #Y Split

    
    
    #Train-Test Split
    X_train, X_test, Y_train, Y_test = splitTrTt(X,Y)
    Y_pred_train,Y_pred_test = lnrReg(X_train, Y_train,X_test)
    Y_pred_train_dtr,Y_pred_test_dtr = DecTreReg(X_train, Y_train,X_test)
    Y_pred_train_LinGCV, Y_pred_test_LinGCV = LinGridCV(X_train, Y_train,X_test)
    Y_pred_train_dtrGCV, Y_pred_test_dtrGCV = dtrGridCV(X_train, Y_train,X_test)
    Y_pred_train_rfr,Y_pred_test_rfr = RanForReg(X_train, Y_train,X_test)
    Y_pred_train_rfcGCV,Y_pred_test_rfcGCV = rfrGridCV(X_train, Y_train,X_test)
    
    
    a = []
    c = [(Y_pred_train,Y_pred_test),(Y_pred_train_dtr,Y_pred_test_dtr),
         (Y_pred_train_LinGCV, Y_pred_test_LinGCV),(Y_pred_train_dtrGCV, Y_pred_test_dtrGCV),
         (Y_pred_train_rfr,Y_pred_test_rfr),(Y_pred_train_rfcGCV,Y_pred_test_rfcGCV)]

    
    #For loop to iterate over the models for calculation of metrics
    from sklearn import metrics
    for i,j in c: 
        metlst = [format(metrics.mean_absolute_error(Y_train,i),'.5f'),
                  format(metrics.mean_absolute_error(Y_test,j),'.5f'),
                  format(metrics.mean_squared_error(Y_train,i),'.5f'),
                  format(metrics.mean_squared_error(Y_test,j),'.5f'),
                  format(np.sqrt(metrics.mean_squared_error(Y_train,i)),'.5f'),
                  format(np.sqrt(metrics.mean_squared_error(Y_test,j)),'.5f'),
                  format(metrics.r2_score(Y_train,i),'.5f'),
                  format(metrics.r2_score(Y_test,j),'.5f')]
        a.append(metlst)

    #DataFrame for storing all the measuring values from models    
    b = pd.DataFrame(a,
                     columns=['MAE_Train', 'MAE_Test','MSE_Train', 'MSE_Test','RMSE_Train', 'RMSE_Test','R2_Train', 'R2_Test'],
                     index = ['Linear Regression','DecisionTree Reg','Linear Grid SearchCV','DecisionTree Reg GSCV',
                              'RandomForest Reg', 'RandomForest Reg GSCV']).astype('float')
    return b
```
